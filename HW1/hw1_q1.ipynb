```python
# Importing required modules

import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
# Question 1(a)

d = [(2**i) for i in range(11)]
d = np.array(d)

def euclidean_distance(vec1, vec2):
    sum = 0
    for i in range (0, len(vec1)):
	    sum += (vec1[i]-vec2[i])**2
    return sum

mean = np.array([])
std = np.array([])

for dim in d:
    sample = np.random.rand(100, dim)
    dist = np.zeros(50*99)
    index = 0
    for i in range(100):
        for j in range(i+1,100):
            dist[index] = euclidean_distance(sample[i], sample[j])
            index += 1
    mean = np.append(mean, np.mean(dist))
    std = np.append(std, np.std(dist))

plt.plot(d,mean,'o-',label='Average')
plt.plot(d,std,'o-',label='Standard deviation')
plt.xlabel('Dimension')
plt.ylabel('Mean/Std. deviation')
plt.title('Euclidian squared distance')
plt.legend()
plt.show()
```
